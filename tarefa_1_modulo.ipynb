Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
```

```
import re
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

padrao_selic = (r'\d+,\d+%')
padrao_ano = (r'\d+ anos?')

correspondencias_selic = re.search(padrao_selic, noticia)
correspondencias_ano = re.search(padrao_ano, noticia)

selic = correspondencias_selic.group() if correspondencias_selic else None
ano = correspondencias_ano.group() if correspondencias_ano else None

print('Taxa Selic:', selic)
print('Ano:', ano)
```
